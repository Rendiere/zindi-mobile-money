## Calculate Average Lat/Lon for Region, district and ward

### Region

#### Use the 2 largest datasets and compare distance

```python
from math import radians, cos, sin, asin, sqrt

def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371 # Radius of earth in kilometers. Use 3956 for miles
    return c * r


def calc_distance(row):
    lat1, lon1, lat2, lon2 = row
    return haversine(lon1, lat1, lon2, lat2)

agg1 = mm_df.groupby(['region']).agg({'latitude': 'mean', 'longitude': 'mean'})

agg2 = ppp_df.groupby(['region']).agg({'latitude': 'mean', 'longitude': 'mean'})

agg_coords_df = pd.merge(agg1, agg2, left_index=True, right_index=True, suffixes=('_mm', '_ppp'))

agg_coords_df['distance_diff_km'] = agg_coords_df.apply(calc_distance,axis=1)
agg_coords_df.sort_values(by='distance_diff_km', ascending=False)
```

**Insight**

The maximum distance between the two aggregated points is 38 km, which is acceptable on a regional level. 

```python
# Save regional lat lons
mm_df.groupby(['region']).agg({'latitude': 'mean', 'longitude': 'mean'}).to_csv('../../data/processed/regions_coords.csv')
```

### District

```python
agg1 = mm_df.groupby(['district']).agg({'latitude': 'mean', 'longitude': 'mean'})
agg2 = ppp_df.groupby(['district']).agg({'latitude': 'mean', 'longitude': 'mean'})

print(agg1.shape, agg2.shape)

agg_coords_df = pd.merge(agg1, agg2, left_index=True, right_index=True, suffixes=('_mm', '_ppp'), how='outer')

agg_coords_df['distance_diff_km'] = agg_coords_df.apply(calc_distance,axis=1)

agg_coords_df.sort_values(by='distance_diff_km', ascending=False)
```

Show the districts not found in both datasets

```python
index_mismatches = set(agg1.index) - set(agg2.index)
index_mismatches
```

50km difference on a district level is actually a bit tight, but lets go with it. 

```python
mm_df.groupby(['district']).agg({'latitude': 'mean', 'longitude': 'mean'}).to_csv('../../data/processed/district_coords.csv')
```

### Ward

```python
agg1 = mm_df.groupby(['ward']).agg({'latitude': 'mean', 'longitude': 'mean'})

agg2 = ppp_df.groupby(['ward']).agg({'latitude': 'mean', 'longitude': 'mean'})

agg_coords_df = pd.merge(agg1, agg2, left_index=True, right_index=True, suffixes=('_mm', '_ppp'))

agg_coords_df['distance_diff_km'] = agg_coords_df.apply(calc_distance,axis=1)

agg_coords_df.sort_values(by='distance_diff_km', ascending=False).head(10)
```

Now it starts getting untrustworthy...

```python
agg_coords_df.sort_values(by='distance_diff_km').head(5)
```

Some wards have a very low difference though. Lets use the average of the two aggregates.

```python
agg_coords_df['lat'] = agg_coords_df[['latitude_mm', 'latitude_ppp']].mean(axis=1)
agg_coords_df['lon'] = agg_coords_df[['longitude_mm', 'longitude_ppp']].mean(axis=1)

ward_df = agg_coords_df[['lat','lon']]
ward_df.to_csv('../../data/processed/ward_coords.csv')
ward_df.head()
```
